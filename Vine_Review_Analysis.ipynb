```python
#dependencies
import pandas as pd
```

```python
#read in csv file
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
#filter df for only total votes > 20
vine_20 = vine_df[vine_df.total_votes>=20]
vine_20.head()
```

```python
#filter for helpful votes divided total votes => 50%
vine_greater_50 = vine_20[vine_20.helpful_votes/vine_20.total_votes >=.50]
vine_greater_50
```

```python
vine_Y_df = vine_greater_50[vine_greater_50.vine == 'Y']
vine_Y_df
```

```python
vine_N_df = vine_greater_50[vine_greater_50.vine == 'N']
vine_N_df
```

```python
#total number of reviews
vine_df.count()
```

```python
#total number of 5 star rated reviews
vine_df[vine_df.star_rating==5].count()
```

```python
#percentage of 5-star reviews for paid and unpaid
num_paid = vine_Y_df.count()
percent_paid = num_paid/2639935
percent_paid
```

```python
#percentage of 5-star reviews for unpaid
num_unpaid =  vine_N_df.count()
percent_unpaid = num_unpaid/2639935
percent_unpaid
```
